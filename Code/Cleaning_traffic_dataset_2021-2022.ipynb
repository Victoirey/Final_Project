```python
velo_TS=velo[['date_comptage', 'comptage_horaire']].groupby(by='date_comptage').sum()
velo_TS
```

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Value', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(velo_TS, x=velo_TS.index, y=velo_TS.comptage_horaire, title='Comptage journalier 2021-2022')  
```
